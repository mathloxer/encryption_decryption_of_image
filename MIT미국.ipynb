```python
import os
from collections import Counter

# 데이터 경로 지정
data_path = "/content/MS-SNSD/clean_train"  # 오디오 파일 경로 수정

# 파일 이름에서 _ 뒤의 숫자 추출
def count_suffix_numbers(data_path):
    file_list = os.listdir(data_path)
    suffix_counts = Counter()

    for file_name in file_list:
        if file_name.endswith(".wav"):  # .wav 파일만 처리
            try:
                suffix = file_name.split("_")[1].split(".")[0]
                suffix_counts[suffix] += 1
            except IndexError:
                pass  # 잘못된 파일 형식 예외 처리

    return suffix_counts

# 함수 실행
suffix_counts = count_suffix_numbers(data_path)

# 결과 출력
for suffix, count in sorted(suffix_counts.items()):
    print(f"{suffix}: {count}개")
```

```python
import os
import librosa
import numpy as np
import soundfile as sf
from joblib import Parallel, delayed
```

```python

# 데이터셋 경로 설정
DATASET_PATH = "/content/drive/MyDrive/MIT/"
TEMP_OUTPUT_PATH = "/content/cutted_Audio/"  # 임시 저장 경로
FINAL_OUTPUT_PATH = "/content/drive/MyDrive/MIT/final_dataset/" # 최종 저장 경로
os.makedirs(TEMP_OUTPUT_PATH, exist_ok=True)
os.makedirs(FINAL_OUTPUT_PATH, exist_ok=True)

# 음성 길이를 3초로 맞추는 함수
def adjust_audio_length(file_path, sr=16000, fixed_length=3):
    try:
        # 파일 로딩
        y, _ = librosa.load(file_path, sr=sr, duration=fixed_length)

        # 3초로 고정된 샘플 수 계산
        fixed_sample_count = sr * fixed_length

        # 패딩 (만약 3초 미만일 경우)
        if len(y) < fixed_sample_count:
            pad_width = fixed_sample_count - len(y)
            y = np.pad(y, (0, pad_width), mode='constant')

        return y
    except Exception as e:
        print(f"Error adjusting audio length for {file_path}: {e}")
        return None

# 3초로 맞춘 파일 저장 함수
def save_adjusted_audio(file_path, temp_output_path, sr=16000, fixed_length=3):
    adjusted_audio = adjust_audio_length(file_path, sr=sr, fixed_length=fixed_length)
    if adjusted_audio is not None:
        # 파일 경로에 따라 하위 폴더 생성
        folder_name = os.path.basename(os.path.dirname(file_path))
        folder_output_path = os.path.join(temp_output_path, folder_name)
        os.makedirs(folder_output_path, exist_ok=True)

        output_file_path = os.path.join(folder_output_path, os.path.basename(file_path))
        sf.write(output_file_path, adjusted_audio, sr)
        return output_file_path
    return None

# 파일 처리
FILE_PATHS = [os.path.join(DATASET_PATH, str(i)) for i in range(2, 11)]  # 2~10명 섞인 폴더

print("Adjusting audio length and saving...")

for folder_path in FILE_PATHS:
    print(f"Processing folder: {folder_path}")
    files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.wav')]

    adjusted_files = Parallel(n_jobs=4)(
        delayed(save_adjusted_audio)(file, TEMP_OUTPUT_PATH) for file in files
    )
    adjusted_files = [f for f in adjusted_files if f is not None]
```

```python
import os
import shutil
import pandas as pd

# 입력 및 출력 디렉토리 설정
root_directory = "/content/cutted_Audio"  # 2, 3, ..., 10 폴더가 포함된 디렉토리
combined_directory = "/content/combined_dataset/"  # 합쳐진 데이터셋 디렉토리
os.makedirs(combined_directory, exist_ok=True)

# 라벨 데이터 저장용 리스트
label_data = []

# 폴더 합치기
for label in range(2, 11):  # 폴더 이름 2 ~ 10
    folder_path = os.path.join(root_directory, str(label))

    if not os.path.exists(folder_path):
        print(f"폴더 {folder_path}가 존재하지 않습니다. 스킵합니다...")
        continue

    print(f"폴더 {folder_path} 처리 중...")

    for file in os.listdir(folder_path):
        if file.endswith(".wav"):
            src_path = os.path.join(folder_path, file)
            dst_filename = f"{label}_{file}"
            dst_path = os.path.join(combined_directory, dst_filename)

            # 파일 복사
            shutil.copy2(src_path, dst_path)

            # 라벨 추가
            label_data.append({
                "file_name": dst_filename,
                "label": label-2
            })

# 라벨 데이터를 CSV 파일로 저장
labels_df = pd.DataFrame(label_data)
labels_csv_path = os.path.join(combined_directory, "labels.csv")
labels_df.to_csv(labels_csv_path, index=False)
print(f"폴더 합치기 및 라벨링 완료. 라벨 데이터는 {labels_csv_path}에 저장되었습니다.")
```

```python
import librosa
import torch
from transformers import Wav2Vec2Processor

combined_directory = "/content/combined_dataset"
labels_csv_path = "/content/combined_dataset/labels.csv"

# Processor 로드
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base")

# 전처리 함수
def preprocess_audio(file_path, target_sr=16000):
    audio, sr = librosa.load(file_path, sr=target_sr)
    return audio

# 데이터 전처리 및 텐서 변환
def process_and_label_data(combined_dir, labels_csv_path, processor, target_sr=16000):
    labels_df = pd.read_csv(labels_csv_path)
    dataset = []

    for _, row in labels_df.iterrows():
        file_path = os.path.join(combined_dir, row["file_name"])
        label = row["label"]

        try:
            audio = preprocess_audio(file_path, target_sr)
            inputs = processor(audio, sampling_rate=target_sr, return_tensors="pt", padding=True)
            dataset.append((inputs.input_values[0], torch.tensor(label)))
        except Exception as e:
            print(f"오류 발생: {file_path} - {e}")

    return dataset

# 데이터 전처리 실행
dataset = process_and_label_data(combined_directory, labels_csv_path, processor)
print(f"전처리 완료. 총 {len(dataset)}개의 데이터가 준비되었습니다.")
```

```python
from sklearn.model_selection import train_test_split

# 데이터 셔플 및 분할
train_data, temp_data = train_test_split(dataset, test_size=0.2, random_state=42)
test_data, val_data = train_test_split(temp_data, test_size=0.5, random_state=42)

print(f"Train 데이터: {len(train_data)}, Test 데이터: {len(test_data)}, Validation 데이터: {len(val_data)}")
```

```python
from transformers import Wav2Vec2ForSequenceClassification, Trainer, TrainingArguments
from torch.utils.data import Dataset

# PyTorch Dataset 클래스 정의
class AudioDataset(Dataset):
    def __init__(self, dataset):
        self.dataset = dataset

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        return {
            "input_values": self.dataset[idx][0],
            "labels": self.dataset[idx][1],
        }

# 데이터셋 생성
train_dataset = AudioDataset(train_data)
test_dataset = AudioDataset(test_data)
val_dataset = AudioDataset(val_data)

# 모델 로드

model = Wav2Vec2ForSequenceClassification.from_pretrained(
    "facebook/wav2vec2-base",
    num_labels=9  # 클래스 개수
)


training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",  # 평가 전략
    save_strategy="epoch",       # 저장 전략을 평가 전략과 일치
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    num_train_epochs=3,
    save_total_limit=2,
    logging_dir="./logs",
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    eval_accumulation_steps=10
)
# Trainer 생성
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=processor,
)
!export CUDA_LAUNCH_BLOCKING=1
```

```python
from evaluate import load

# 정확도 메트릭 로드
accuracy_metric = load("accuracy")

# 평가 메트릭 함수 정의
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = logits.argmax(axis=-1)
    return accuracy_metric.compute(predictions=predictions, references=labels)

# TrainingArguments 설정
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",  # 평가 주기: epoch
    save_strategy="epoch",        # 저장 주기: epoch
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    num_train_epochs=3,
    save_total_limit=2,
    logging_dir="./logs",
    load_best_model_at_end=True,
    metric_for_best_model="eval_accuracy",  # 정확도 기준으로 최적 모델 선택
    greater_is_better=True,  # 높은 값이 더 좋은 것으로 간주
)

# Trainer 초기화
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=processor,
    compute_metrics=compute_metrics,  # 평가 메트릭 함수 추가
)

# 학습 시작
trainer.train()
```

```python
import torch
from transformers import Wav2Vec2ForSequenceClassification, Wav2Vec2Processor
import librosa

# 저장된 모델과 프로세서 로드
model = Wav2Vec2ForSequenceClassification.from_pretrained("./fine_tuned_model")
processor = Wav2Vec2Processor.from_pretrained("/content/drive/MyDrive/MIT/final_model")

# 예측 함수 정의
def predict_audio_label(audio_file, model, processor, target_sr=16000):
    """
    오디오 파일을 입력으로 받아, 학습된 Wav2Vec2 모델로 예측.
    """
    try:
        # 오디오 파일 로드 및 샘플링 속도 변환
        audio, sr = librosa.load(audio_file, sr=target_sr)

        # Processor를 사용하여 모델 입력 준비
        inputs = processor(audio, sampling_rate=target_sr, return_tensors="pt", padding=True)

        # 모델 예측
        model.eval()
        with torch.no_grad():
            logits = model(inputs.input_values).logits
            predicted_class = logits.argmax(dim=-1).item()

        return predicted_class
    except Exception as e:
        print(f"오디오 파일 예측 중 오류 발생: {e}")
        return None

# 예제 오디오 파일로 예측 수행
example_audio = "/content/Test 1.m4a"  # 예측할 오디오 파일 경로
predicted_label = predict_audio_label(example_audio, model, processor)

if predicted_label is not None:
    print(f"예측된 클래스: {predicted_label}")
else:
    print("예측에 실패했습니다.")
```
