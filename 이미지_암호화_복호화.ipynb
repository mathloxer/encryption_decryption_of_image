```python
from PIL import Image
import matplotlib.pyplot as plt
import os
# from google.colab import files # 파일 업로드는 이제 필요 없으므로 주석 처리

def view_multiple_ppm_images_in_colab(image_paths):
    """
    지정된 여러 PPM 이미지들을 Colab 노트북 내에 한 번에 표시합니다.
    """
    num_images = len(image_paths)
    if num_images == 0:
        print("표시할 이미지가 없습니다.")
        return

    # 이미지들을 나란히 표시하기 위한 서브플롯 설정
    plt.figure(figsize=(5 * num_images, 5)) # 이미지 개수에 따라 전체 Figure 크기 조절

    for i, image_path in enumerate(image_paths):
        if not os.path.exists(image_path):
            print(f"오류: '{image_path}' 파일을 Colab 환경에서 찾을 수 없습니다. 경로를 확인해주세요.")
            continue

        try:
            with Image.open(image_path) as img:
                plt.subplot(1, num_images, i + 1) # 1행, num_images 열, i+1번째 서브플롯
                plt.imshow(img)
                plt.title(os.path.basename(image_path)) # 파일명을 제목으로
                plt.axis('off') # 축 정보 숨기기

        except Exception as e:
            print(f"오류: '{image_path}' 파일을 여는 중 문제가 발생했습니다: {e}")

    plt.tight_layout() # 서브플롯 간의 간격 자동 조정
    plt.show() # 이미지 출력

if __name__ == "__main__":
    # C 코드에서 생성된 PPM 파일들의 이름을 여기에 정확히 입력합니다.
    # Colab에 이 파일들을 미리 업로드해두어야 합니다.
    ppm_files_to_view = [
        "/content/original_generated_gradient.ppm",
        "/content/encrypted_gradient_image.ppm",
        "/content/decrypted_gradient_image.ppm"
    ]

    print("미리 업로드된 PPM 이미지들을 로드하여 표시합니다.")
    view_multiple_ppm_images_in_colab(ppm_files_to_view)

    print("\n모든 PPM 파일 표시 완료.")
```
